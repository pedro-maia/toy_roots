```python
import numpy as np
import matplotlib.pyplot as plt

# v1: line y = a x + b
a, b = 2.1, -4.1  # change these

# root of y=0
if abs(a) < 1e-12:
    root = None if abs(b) > 1e-12 else "infinitely many"
else:
    root = -b / a

x = np.linspace(-10, 10, 400)
y = a*x + b

plt.plot(x, y)
plt.axhline(0); plt.axvline(0)
if isinstance(root, float):
    plt.scatter([root], [0])
plt.title(f"y={a}x+{b} | root={root}")
plt.show()

print("root =", root)
```
